### **numpy가 왜 필요할까?**

####문제1
*   1부터 20까지 3씩 증가되는 수열(리스트)을 생성하시오.
*   리스트의 모든 항목 값에 10을 더하시오.
*   리스트의 모든 항목 값에 2를 곱하시오.

```python
x = list(range(1, 20, 3)) # range는 정수만 생성 가능
x
```

```python
for i in range(len(x)):
  x[i] += 10
x
```

```python
for i in range(len(x)):
  x[i] *= 2
x
```

####문제2
*   0.2부터 2.5까지 0.3씩 증가되는 수열(리스트)을 생성하시오.
*   리스트의 모든 항목 값에 10을 더하시오.
*   리스트의 모든 항목 값에 2를 곱하시오.

```python
import numpy as np
# range는 실수 생성 불가
x = np.arange(0.5, 2.5, 0.3) # arange는 배열을 반환함
x
```

```python
x += 10
x
```

```python
x *= 2
x
```

### **numpy 사용하지 않는 경우와 사용하는 경우 비교**
numpy 사용하지 않는 경우

```python
중간 = [90, 80, 70, 60, 50]
기말 = [80, 70, 60, 50, 40]
최종 = []

#최종 = 중간 + 기말 # 연결됨
for i in range(5):
  hap = (중간[i] + 기말[i]) / 2
  최종.append(hap)

print(최종)

for i in range(len(최종)):
  최종[i] += 7

print("가산점 적용 후=", 최종)
```

numpy 사용하는 경우

```python
중간 = [90, 80, 70, 60, 50]
기말 = [80, 70, 60, 50, 40]

mid = np.array(중간)
final = np.array(기말)

최종 = (mid + final) / 2
print(최종)

최종 += 7
print("가산점 적용 후=", 최종)
```

### **np.arange()를 이용한 Sin함수 그리기**

```python
import matplotlib.pyplot as plt

x = np.arange(0, 3.14*2, 0.1)

y = np.sin(x)

plt.plot(x,y)
plt.show()
```

### **np.arange()를 이용한 Sin, Cos함수 그리기**

```python
x = np.arange(0, 3.14*2, 0.1)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label='sine')
plt.plot(x,y2, label='cosine')
plt.title("sine & cosine graph")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.grid(True)
plt.legend()
plt.show()
```

### **np.linspace()를 이용한 Sin, Cos함수 그리기**

```python
x = np.linspace(-np.pi, np.pi, 100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1, label='sine')
plt.plot(x,y2, label='cosine')
plt.title("sine & cosine graph")
plt.xlabel("X value")
plt.ylabel("Y value")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# np.linspace(시작값, 종료값, 데이터 생성 개수(등급의 개수))
a = np.linspace(1, 10, 10) # linspace는 종료값까지 포함
a
```

```python
a = np.linspace(1, 25, 30)
a
```
