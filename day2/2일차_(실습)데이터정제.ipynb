```python
#결측치 제거하기

score2 = pd.read_excel('/content/성적2.xlsx')
score2
```

```python
#DataFrame의 행/열 정보 확인

score2.info()
```

```python
#결측값 제거하기
#df.dropna(subset=['열이름1','열이름2'])
#subset=['열이름'] : 특정열에 NaN이 존재하는 행만 선택하여 삭제
#subset=['열이름1','열이름2'] : 열이름1 또는 열이름2에 NaN이 있는 모든 행 삭제

score2.dropna(subset=['출석'])
```

```python
#결측값 제거하기
#변수명.dropna(subset=['열이름1','열이름2'])
#subset=['열이름'] : 특정열에 NaN이 존재하는 행만 선택하여 삭제
#subset=['열이름1','열이름2'] : 열이름1 또는 열이름2에 NaN이 있는 모든 행 삭제

score2.dropna(subset=['프로젝트', '실험'])
```

```python
#행과 열에 존재하는 결측치를 선택하여 삭제
#변수명.dropna(axis=0, how='any', thresh=개수, inplace=True)
#axis=0 또는 1 : 행과 열을 선택하여 삭제
#how='any' : 결측치의 포함 정도에 따라 삭제
#thresh=개수 : 유효한 데이터가 존재하는 '개수'이상만 남기고 삭제
#inplace=True : 원본 데이터에 반영

score2.dropna(axis=0) # 맨 아래 행부터 삭제
```

```python
# 행, 결측치 포함 정도에 따라 삭제

score2.dropna(axis=0, how='any')
```

```python
# 행, 결측치 포함 정도에 따라 삭제
#all : 모두 포함하면 행/열 삭제

score2.dropna(axis=0, how='all')
```

```python
#행과 유효한 데이터가 존재하는 '개수'이상만 남기고 삭제

score2.dropna(axis=0, thresh=4)
```

```python
#결측치 치환하기

score3 = pd.read_excel('/content/성적3.xlsx')
score3
```

```python
#앞/뒤 행의 중간값으로 치환
#변수명.interpolate()

score3.interpolate()
```

```python
#앞 행의 값으로 치환
#변수명.fillna(method='ffill')

score3.fillna(method='ffill')
```

```python
#뒤 행의 값으로 치환
#변수명.fillna(method='bfill')

score3.fillna(method='bfill')
```

```python
#Table에 있는 전체 NaN을 동일값으로 변경
#변수명.fillna(값)

score3.fillna(50)
```

```python
#특정열에 특정값을 치환하고자 하는 경우
#변수명.fillna({'열이름':값})

score3.fillna({'출석':10, '프로젝트':20})
```

```python
# 특정열의 평균값으로 대체
#변수명.fillna(변수명['열이름'].mean())

score3.fillna(score3['출석'].mean())
```

```python
#이상치 확인 및 치환하기

score4 = pd.read_excel('/content/성적4.xlsx')
score4
```

```python
# 이상치 확인하기
#변수명['열이름'].value_counts()

score4['성별'].value_counts()
```

```python
#이상치 데이터 치환하기
#변수명['열이름']=변수명['열이름'].replace('찾는 데이터','변환할 데이터')


score4['성별'] = score4['성별'].replace(3, 2)
score4
```

```python
#'열이름'을 기준으로 중복되는 행 검출
#변수명.duplicated(['열이름'])

score4.duplicated(['ID'])
```

```python
#'열이름'을 기준으로 중복 데이터 행 삭제
#변수명.drop_duplicates(['열이름'])

score4.drop_duplicates('ID')
```

```python
#2개 열의 데이터가 일치하는 행 삭제
#변수명.drop_duplicates(subset = ['열이름','열이름'],  keep=False)
#keep=" " : 데이터를 유지할 행 설정
#기본은 첫 번째 데이터를 유지
#"last" : 마지막 데이터 유지
#False : 모두 삭제

score4.drop_duplicates(subset=['ID', '성별'], keep=False, inplace=True)
score4
```
