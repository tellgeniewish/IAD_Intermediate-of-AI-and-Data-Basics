```python
import requests

url = "http://openapi.seoul.go.kr:8088/4d434d57546a6164393571766a6e42/json/SebcCollegeInfoKor/1/1000/"
# http://openapi.seoul.go.kr:8088/4d434d57546a6164393571766a6e42/json/SebcCollegeInfoKor/1/1000
# 즉, 10000만 건일 경우, 코랩에서 에러 확인X
# 주소 복붙 후 새 창에서 에러 확인 ㄱㄴ

response = requests.get(url)
print(response.content) # 바이너리 형태로 가져온다

data = response.json()
data
```
