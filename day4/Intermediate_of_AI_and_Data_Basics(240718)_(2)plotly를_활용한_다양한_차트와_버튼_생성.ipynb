```python
px.bar(x=혼인건수['2022'], y=혼인건수['자치구별'], color=혼인건수['자치구별'], orientation='h', text_auto=True)
```

```python
혼인건수2022 = 혼인건수.sort_values('2022', ascending=False).head()
혼인건수2022
```

```python
px.pie(values=혼인건수2022['2022'], names=혼인건수2022['자치구별'])
```

```python
fig = px.pie(values=혼인건수2022['2022'], names=혼인건수2022['자치구별'])
# hole에는 0~1사이의 실수값을 넣어줘야 한다
fig.update_traces(hole=.3) # .3은 0.3이라는 의미
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=혼인건수['자치구별'], y=혼인건수['2020'], name='2020'))
fig.add_trace(go.Bar(x=혼인건수['자치구별'], y=혼인건수['2022'], name='2022'))
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=혼인건수['자치구별'], y=혼인건수['2020'], name='2020'))
fig.add_trace(go.Scatter(x=혼인건수['자치구별'], y=혼인건수['2022'], mode='lines', name='2022'))
fig.show()
```

```python
fig.write_html('plotly차트.html')
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=혼인건수['자치구별'], y=혼인건수['2020'], name='2020'))
fig.add_trace(go.Scatter(x=혼인건수['자치구별'], y=혼인건수['2022'], mode='lines', name='2022'))

fig.update_layout(
    updatemenus=[
        dict(
            type='buttons',
            buttons=list([
                dict(label="모두 그리기", method='update', args=[{'visible':[True, True]}]),
                dict(label="2020 모두 그리기", method='update', args=[{'visible':[True, False]}]),
                dict(label="2022 모두 그리기", method='update', args=[{'visible':[False, True]}])
                ])
            )
        ]
    )
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=혼인건수['자치구별'], y=혼인건수['2020'], mode='lines', name='2020'))
fig.add_trace(go.Scatter(x=혼인건수['자치구별'], y=혼인건수['2022'], mode='lines', name='2022'))

fig.update_layout(
    updatemenus=[
        dict(
            type='buttons',
            buttons=list([
                dict(label="line 차트", method='restyle', args=['type', 'scatter']),
                dict(label="Bar 차트", method='restyle', args=['type', 'bar'])
                ])
            )
        ]
    )
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=혼인건수['자치구별'], y=혼인건수['2020'], name='2020'))
fig.add_trace(go.Bar(x=혼인건수['자치구별'], y=혼인건수['2022'], name='2022'))

fig.update_layout(
    updatemenus=[
        dict(
            type='buttons',
            buttons=list([
                dict(label="Bar 차트", method='restyle', args=['type', 'bar']),
                dict(label="line 차트", method='restyle', args=['type', 'scatter'])
                ])
            )
        ]
    )
fig.show()
```
